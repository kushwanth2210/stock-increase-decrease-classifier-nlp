```
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```
cv=CountVectorizer(ngram_range=(2,2))
traindataset=cv.fit_transform(NEWS_HEADLINES)
classifier=RandomForestClassifier(n_estimators=200,criterion='entropy')
classifier.fit(traindataset,train['Label'])
```

```
test_transformation=[]
for row in range(0,len(test.index)):
  test_transformation.append(' '.join(str(x) for x in test.iloc[row,0:25]))
test_ds=cv.transform(test_transformation)
predictions=classifier.predict(test_ds)
```

```
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
c_m=confusion_matrix(test['Label'],predictions)
c_r=classification_report(test['Label'],predictions)
acc_score=accuracy_score(test['Label'],predictions)
print(c_m)
print(c_r)
print(acc_score)
```
